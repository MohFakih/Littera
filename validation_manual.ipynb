```python
from tensorflow.keras.models import load_model
import numpy as np
import matplotlib.pyplot as plt
from imageLoad import getPathToAllImages, sampleImages
```

```python
model = load_model(".\\models\\30.h5")
tensorboardLogs = ".\\Logs"
```

```python
pathToAllImages = getPathToAllImages()
```

```python
imgsRaw = sampleImages(pathToAllImages, 1, 256, 256)
imgs = np.uint8(255*imgsRaw)
outt = model.predict(imgsRaw)
o1 = 255*outt[0]
plt.imshow(imgs[0].reshape((256, 256)), cmap='gray', vmin=0, vmax=256)
plt.show()
plt.imshow(o1.reshape((256, 256)), cmap='gray', vmin=0, vmax=256)
```
