```python
from tensorflow.keras import Sequential
from tensorflow.keras.models import load_model
import matplotlib.pyplot as plt
import numpy as np
autoencoder = load_model(".\\models\\35.h5")
```

```python
generator = Sequential()
G = autoencoder.layers[5:]
for layer in G:
    generator.add(layer)
generator.build((None, 16, 16, 1))
generator.summary()
```

```python
outt = generator.predict(np.random.random((1, 16, 16, 1), ))
plt.imshow(outt[0].reshape((256, 256)), cmap='gray')

```
